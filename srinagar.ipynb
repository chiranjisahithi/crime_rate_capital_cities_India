```python
import pandas as pd
import numpy as np
from bokeh.plotting import output_file, show, figure, ColumnDataSource
import matplotlib.pyplot as plt 
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
df=pd.read_csv("https://code.swecha.org/CivicTech/analysis-on-survillance-in-indian-cities/-/raw/Challengers/srinagar.csv",encoding='cp1252')
pd.set_option('display.max_rows',None)
df
```

```python
output_file("CRIMESRI.html")
year=[]
cameras=[]
l1=[]
for i in range(0,11):
    l1.append(df.year[i])
    cameras.append(df.Camerafor1000_people[i])
for i in l1 :
    year.append(str(i))
p=figure(x_range=year,title="CRIME IN SRINAGAR",y_axis_label="NUMBER OF CAMERAS",x_axis_label="YEAR")
p.vbar(x=year,top=cameras,width=0.6,color="firebrick")
show(p)
```

```python
output_file("CRIMESRI.html")
crime=[]
cameras=[]
l1=[]
for i in range(0,11):
    l1.append(df.Camera_for_1000_people[i])
    crime.append(df.Crime_index[i])
for i in l1 :
    cameras.append(str(i))
p=figure(x_range=cameras,title="CRIME IN DELHI",y_axis_label="Crime rate",x_axis_label="Cameras")
p.vbar(x=cameras,top=crime,width=0.6,color="firebrick")
show(p)
```

```python
source=ColumnDataSource(df)
p=figure(plot_width=400,plot_height=400,title='SURVEILLANCE IN SRINAGAR',x_axis_label='YEAR',y_axis_label='Number of Cameras')
output_file("linecamsri.html")
p.line(x="year", y="Camerafor1000_people", source=source,line_width=4,color="red")
p.square(x="year" , y="Camerafor1000_people", source=source,color="yellow")
hover = HoverTool()
hover.tooltips=[('Year','@year'),('Number of Cameras','@Camerafor1000_people'),('RESIDENTIAL ROBBERY','@RESIDENTIALPREMISES_Robbery'),('RESIDENTIAL BURGLARY','@RESIDENTIALPREMISES_Burglary'),('RESIDENTAIL THEFT','@RESIDENTIALPREMISES_Theft'),('HIGHWAY ROBBERY','@HIGHWAYS_Robbery'),('HIGHWAY THEFT','@HIGHWAYS_Theft'),('THEFT IN RAILWAYS','@RAILWAYS_Theft'),('N BANKS DACOITY I','@BANKS_Dacoity'),('ROBBERY IN BANKS','@BANKS_Robbery'),('THEFT IN BANKS','@BANKS_Theft'),('CRIME ARTE','@Crime_index')]
p.add_tools(hover)
show(p)
```

```python
from bokeh.plotting import figure,output_file,show

output_file("sriSCA.html")
p=figure(plot_width=1100,plot_height=500,title="CRIME IN DELHI",y_axis_label="values",x_axis_label="year")

p.circle(df["year"],df["RESIDENTIALPREMISES_Robbery"],size=12,line_color='green',fill_color='orange',fill_alpha=1,legend_label="RESIDENTIAL ROBBERY")

p.square(df["year"],df["RESIDENTIALPREMISES_Theft"],size=10,line_color='green',fill_color='yellow',fill_alpha=1,legend_label="RESIDENTIAL THEFT")

p.triangle(df["year"],df["HIGHWAYS_Robbery"],size=14,line_color='cyan',fill_color='blue',fill_alpha=1,legend_label="HIGHWAY ROBBERY")


p.diamond(df["year"],df["HIGHWAYS_Theft"],size=15,line_color='black',fill_color='red',fill_alpha=1,legend_label="THEFT IN HIGHWAY")

p.hex(df["year"],df["Camerafor1000_people"],size=15,line_color='yellow',fill_color='violet',fill_alpha=1,legend_label="TOTAL CAMERAS")
p.dot(df["year"],df["Crime_index"],size=25,line_color='red',fill_color='black',fill_alpha=1,legend_label="CRIME INDEX")


show(p)

 
```

```python
output_file("pieSRi.html")
year=[]
RPR=[]
for i in range(0,11):
    year.append(str(df.year[i]))
    RPR.append(int(df.Camerafor1000_people[i]))

x=dict(zip(year,RPR))

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'year'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(x)]

p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@year: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='year', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)

```
