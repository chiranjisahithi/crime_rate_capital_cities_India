```python
import pandas as pd
import numpy as np
from bokeh.plotting import output_file, show, figure, ColumnDataSource
import matplotlib.pyplot as plt 
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
df=pd.read_csv("https://code.swecha.org/CivicTech/analysis-on-survillance-in-indian-cities/-/raw/Challengers/ITANAAR.csv",encoding='cp1252')
pd.set_option('display.max_rows',None)
df
```

```python
output_file("CAMERAITAN.html")
year=[]
cameras=[]
l1=[]
for i in range(0,11):
    cameras.append(df.Cameras[i])
    l1.append(df.year[i])
for i in l1 :
    year.append(str(i))
p=figure(x_range=year,title="CAMERAS IN ITANAGAR",y_axis_label="NUMBER OF CAMERAS",x_axis_label="YEAR")
p.vbar(x=year,top=cameras,width=0.6,color="firebrick")
show(p)
```

```python
output_file("CRIMEITAN.html")
year=[]
crime=[]
l1=[]
for i in range(0,11):
    crime.append(df.Crime_index[i])
    l1.append(df.year[i])
for i in l1 :
    year.append(str(i))
p=figure(x_range=year,title="CRIME IN GITANAGAR",y_axis_label="CRIME RATE",x_axis_label="YEAR")
p.vbar(x=year,top=crime,width=0.6,color="cyan")
show(p)
```

```python
from math import pi
import pandas as pd
from bokeh.io import output_file, show
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.plotting import output_notebook
output_notebook()
output_file("itanpie.html")
year=[]
camera=[]
for i in range(0,11):
    year.append(str(df.year[i]))
    camera.append(df.Cameras[i])

x=dict(zip(year,camera))

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'year'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(x)]

p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@year: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='year', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
source=ColumnDataSource(df)
p=figure(plot_width=400,plot_height=400,title='SURVEILLANCE',x_axis_label='YEAR',y_axis_label='Number of Cameras')
output_file("Itan.html")
p.line(x="year", y="Cameras", source=source,line_width=4,color="blue")
p.square(x="year" , y="Cameras", source=source,color="yellow",size=6)
hover = HoverTool()
hover.tooltips=[('Year','@year'),('Number of Cameras','@Cameras'),('Number of residential robbery','@RESIDENTIAl_PREMISESRobbery'),('Number of residential Burglary','@RESIDENTIAL_PREMISESBurglary'),('Number residential of Theft','@RESIDENTIAL_PREMISESTheft'),('Highway Robbery','@HIGHWAYSRobbery'),('Highway theft','@HIGHWAYSTheft'),('Railway theft','@RAILWAYSTheft'),('Bank dacoity','@BANKSDacoity'),('Bank robbey','@BANKSRobbery'),('Bank burglary','@BANKSBurglary'),('Bank Theft','@BANKSTheft'),('Crime Index','@Crime_index')]
p.add_tools(hover)
show(p)
```

```python
from bokeh.plotting import figure,output_file,show

output_file("ITANSCA.html")
p=figure(plot_width=1100,plot_height=500,title="CRIME IN GOA",y_axis_label="values",x_axis_label="year")

p.circle(df["year"],df["RESIDENTIAl_PREMISESRobbery"],size=12,line_color='green',fill_color='orange',fill_alpha=1,legend_label="RESIDENTIAL ROBBERY")

p.square(df["year"],df["RESIDENTIAL_PREMISESTheft"],size=10,line_color='green',fill_color='yellow',fill_alpha=1,legend_label="RESIDENTIAL THEFT")

p.triangle(df["year"],df["HIGHWAYSRobbery"],size=14,line_color='cyan',fill_color='blue',fill_alpha=1,legend_label="HIGHWAY ROBBERY")


p.diamond(df["year"],df["HIGHWAYSTheft"],size=15,line_color='black',fill_color='red',fill_alpha=1,legend_label="THEFT IN HIGHWAY")

p.hex(df["year"],df["Cameras"],size=15,line_color='yellow',fill_color='violet',fill_alpha=1,legend_label="TOTAL CAMERAS")
p.dot(df["year"],df["Crime_index"],size=25,line_color='red',fill_color='black',fill_alpha=1,legend_label="CRIME RATE")

show(p)
```
