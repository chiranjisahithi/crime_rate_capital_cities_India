```python
import pandas as pd
import numpy as np
from bokeh.plotting import output_file, show, figure, ColumnDataSource
import matplotlib.pyplot as plt 
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
df=pd.read_csv("https://code.swecha.org/CivicTech/analysis-on-survillance-in-indian-cities/-/raw/Challengers/hyd.csv",encoding='cp1252')
pd.set_option('display.max_rows',None)
df
```

```python
output_file("CRIMESRI.html")
crime=[]
cameras=[]
l1=[]
for i in range(0,10):
    cameras.append(df.no_ofcameras[i])
    l1.append(df.Crime_index[i])
for i in l1 :
    crime.append(str(i))
p=figure(x_range=crime,title="CRIME IN HYDERABAD",y_axis_label="NUMBER OF CAMERAS",x_axis_label="CRIME RATE")
p.vbar(x=crime,top=cameras,width=0.6,color="orange")
show(p)
```

```python
output_file("CrimeHYD.html")
cases=[]
cameras=[]
l1=[]
for i in range(0,11):
    l1.append(df.Total_Cases[i])
    cameras.append(df.no_ofcameras[i])
for i in l1 :
    crime.append(str(i))
p=figure(x_range=cases,title="CRIME IN HYDERABAD",y_axis_label="NUMBER OF CAMERAS",x_axis_label="CRIME RATE")
p.vbar(x=cases,top=cameras,width=0.6,color="pink")
show(p)
```

```python
from math import pi
import pandas as pd
from bokeh.io import output_file, show
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.plotting import output_notebook
output_notebook()
output_file("HYDPIE1.html")
year=[]
camera=[]
for i in range(0,11):
    year.append(str(df.Year[i]))
    camera.append(df.no_ofcameras[i])

x=dict(zip(year,camera))

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'year'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(x)]

p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@year: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='year', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
source=ColumnDataSource(df)
p=figure(plot_width=400,plot_height=400,title='SURVEILLANCE',x_axis_label='YEAR',y_axis_label='Number of Cameras')
output_file("hyd.html")
p.line(x="Year", y="no_ofcameras", source=source,line_width=4,color="green")
p.square(x="Year" , y="no_ofcameras", source=source,color="yellow",size=6)
hover = HoverTool()
hover.tooltips=[('Year','@Year'),('Number of Cameras','@no_ofcameras'),('Number of residential robbery','@RESIDENTIALPREMISES_Robbery'),('Number of residential Burglary','@RESIDENTIALPREMISES_Burglary'),('Number residential of Theft','@RESIDENTIALPREMISES_Theft'),('Highway Robbery','@HIGHWAYS_Robbery'),('Highway theft','@HIGHWAYS_Theft'),('Railway theft','@RAILWAYS_Theft'),('Bank dacoity','@BANKS_Dacoity'),('Bank robbey','@BANKS_Robbery'),('Bank burglary','@BANKS_Burglary'),('Bank Theft','@BANKS_Theft'),('Total Cases','@Total_Cases')]
p.add_tools(hover)
show(p)
```

```python
from bokeh.plotting import figure,output_file,show

output_file("HYDSCA.html")
p=figure(plot_width=1100,plot_height=500,title="CRIME IN HYDERABAD",y_axis_label="values",x_axis_label="year")

p.circle(df["Year"],df["RESIDENTIALPREMISES_Robbery"],size=12,line_color='green',fill_color='orange',fill_alpha=1,legend_label="RESIDENTIAL ROBBERY")

p.square(df["Year"],df["RESIDENTIALPREMISES_Theft"],size=10,line_color='green',fill_color='yellow',fill_alpha=1,legend_label="RESIDENTIAL THEFT")

p.triangle(df["Year"],df["HIGHWAYS_Robbery"],size=14,line_color='cyan',fill_color='blue',fill_alpha=1,legend_label="HIGHWAY ROBBERY")


p.diamond(df["Year"],df["HIGHWAYS_Theft"],size=15,line_color='black',fill_color='red',fill_alpha=1,legend_label="THEFT IN HIGHWAY")

p.hex(df["Year"],df["no_ofcameras"],size=15,line_color='yellow',fill_color='violet',fill_alpha=1,legend_label="TOTAL CAMERAS")
p.dot(df["Year"],df["Total_Cases"],size=25,line_color='red',fill_color='black',fill_alpha=1,legend_label="TOTAL CASES")


show(p)

 
```
